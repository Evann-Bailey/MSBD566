Question 4: [A vectorization problem]. Compare the manual approach versus a vectorized
approach to find the mean AQI for each Air Quality Category ('Category' in the table).
For the vectorized method, you can use groupby() and mean() method in pandas datatype
(ex: data.groupby('ColumnName') and dataList.mean()). For the manual method, a
regular for-loop can be used. Calculate the time difference between the two approaches and
evaluate.

It is faster to use a vectorized approach to calculate the mean AQI for each air quality category. 

```python
import time

start_manual = time.time()

categories = data["Category"].unique()
manual_means = {}

for cat in categories:
    cat_values = data[data["Category"] == cat]["AQI"]
    manual_means[cat] = cat_values.mean()

end_manual = time.time()
manual_time = end_manual - start_manual

print("Manual method results:", manual_means)
print(f"Manual method time: {manual_time:.4f} seconds")
```

```python
start_vectorized = time.time()

vectorized_means = data.groupby("Category")["AQI"].mean()

end_vectorized = time.time()
vectorized_time = end_vectorized - start_vectorized

print("\nVectorized method results:\n", vectorized_means)
print(f"Vectorized method time: {vectorized_time:.4f} seconds")
```
